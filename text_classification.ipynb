```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('/content/letters.csv')
```

```python
df = df.drop_duplicates()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

# Build Knn model

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop(['label'],axis=1)
y = df['label']
```

```python
vif_data = pd.DataFrame()
vif_data["Feature"] = range(X.shape[1])  # Replace with actual feature names if available
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

# Display VIF values
print(vif_data)
```

```python
scaler_minmax = MinMaxScaler()
x_scaler = scaler_minmax.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size=0.2,random_state=42)
```

```python
Knn_model = KNeighborsClassifier(n_neighbors=5)
model_performance = Knn_model.fit(x_train,y_train)
```

```python
y_pred = model_performance.predict(x_test)
```

```python
accuracy_knn = accuracy_score(y_test,y_pred)
classify_report_knn = classification_report(y_test,y_pred)
```

```python
print("Accuracy:", accuracy_knn)
print("\nClassification Report:\n", classify_report_knn)
```

```python
best_k = 1
best_accuracy = 0
accuracies = []
```

```python
for i in range(1,20) :
  Knn_model = KNeighborsClassifier(n_neighbors=i)
  model_performance = Knn_model.fit(x_train,y_train)
  y_pred = model_performance.predict(x_test)
  accuracy_knn = accuracy_score(y_test,y_pred)
  if accuracy_knn > best_accuracy:
    best_k = i
    best_accuracy = accuracy_knn
```

```python
print(f"Optimal k: {best_k}")
print(f"Accuracy with k={best_k}: {best_accuracy:.4f}")
```

# Neural Networks

```python
df.head()
```

```python
X = X / 255.0
```

```python
model = Sequential([
    # Flatten layer to convert 2D image array to 1D vector
    Flatten(input_shape=(X.shape[1],)),

    # Dense hidden layers
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(),
              loss='binary_crossentropy',  # Use 'categorical_crossentropy' for multi-class
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(x_test, y_test)

# Print the test accuracy
print(f'Test Accuracy: {test_acc:.4f}')
```
